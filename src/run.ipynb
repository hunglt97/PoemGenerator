```
import torch
from transformers import AutoModel, AutoTokenizer

#phobert = AutoModel.from_pretrained("vinai/phobert-base")
GPTtokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base", use_fast=False)
BERTtokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base", use_fast=False)
GPTtokenizer.add_tokens('\n')
```

```
from transformers import pipeline

poem = pipeline('text-generation', model="./poems", tokenizer=GPTtokenizer, config={'max_length':258}, device=0)
```

```
from transformers import RobertaConfig, RobertaForSequenceClassification

model_bert = RobertaForSequenceClassification.from_pretrained("./BERT/models")
```

```
import re

number_of_generated_poems = 5

with open("./poems/results.txt", "w") as f:
  for i in range(number_of_generated_poems):
    tho = poem('<s> ')[0]['generated_text']
    tho = re.sub("<s> ","", tho)
    tho = re.sub(" </s>","\n\n", tho)
    tho_bert = re.sub("\n", " ", tho)
    inputs = BERTtokenizer(tho_bert, return_tensors="pt")
    outputs = model_bert(**inputs)
    logits = outputs.logits
    f.write(tho)
    f.write(f"\n Accuracy, {logits} \n")
    f.write("\n===========================================\n")
```
